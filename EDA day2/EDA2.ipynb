### Problem Statement:

A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
## import the test data too
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
## we use '_append' coz 'append' is deprecated in latest pd version
df=df_train._append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#vwe wil drop user_id using drop function in pd we use axis=1 to remove row wise not  column wise for axis =0
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
## we have used map functionality to convert female val into 0 and male val into 1:)
## we did encoding!
df['Gender']=df['Gender'].map({'F':0, 'M':1})
df.head()
```

```python
# handling feature Age
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head()
```

```python
df.drop('City_Category', axis=1, inplace=True)  ## drop city_category
```

```python
df.head()
```

```python
## Missing Values
df.isnull().sum()
```

```python
## replacing missing values..
df['Product_Category_2'].unique()
```

#### Observation: 
values are discrete and not continous

```python
df['Product_Category_2'].value_counts()
```

```python
## Best way is to replace missing with mode
df['Product_Category_2'].mode()[0]
```

```python
## filling missing values
df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df.isnull().sum()
```

```python
## similarly for product category 3
df['Product_Category_3'].mode()[0]
```

```python
## filling missing values:)
df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
## we will first replace 4+ with only 4
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
df.info()
```

```python
## we have to convert STAY IN CURRENT CITY YEARS(object) into integer
df['Stay_In_Current_City_Years']= df['Stay_In_Current_City_Years'].astype(int)
```

```python
df.info()
```
